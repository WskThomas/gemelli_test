# Présentation AlphaZero

L'idée principale d'AlphaZero est d'obtenir une recherche guidée par un réseau de neurones.
Contrairement à d'autres méthodes (recherche alpha-beta) qui utilise des arbres d'exploration très grands, AlphaZero se base sur une recherche dans un arbre qui est relativement petit. La recherche des actions à faire est faites selon une recherche MCTS (Monte Carlo tree search).

L'entrainement se fait en 3 temps: 
1. Phase 1: Le réseau joue contre lui même afin d'évaluer les différents plateaux (self-play).
2. Phase 2: Le réseau est entrainé afin de coller au mieux a ces évaluation.
2. Phase 3: Le réseau nouvellement entrainé joue contre le réseaux entrainé précédement afin de valider (phase de validation)

Dans les faits le réseaux seul ne choisi pas directement les actions à faire. 
Le réseau guide une Monte Carlo tree search (MCTS) afin de choisir la meilleure action lors du self-play et lors de l'evalution du plateau.

Il s'agit d'une recherche recursive qui permet d'obtenir une action meilleure que le réseau seul.
L'idée de AlphaZero est de faire coller le résultat du réseau à cette exploration qui est plus efficace lors de l'entraînement.
